```python
from langchain_openai import ChatOpenAI
from langchain.agents import initialize_agent, Tool
#from langchain.utilities import PythonREPL
from langchain_experimental.utilities import PythonREPL
```

```python
from langchain_openai import AzureChatOpenAI
```

```python
# LLM for reasoning
llm = AzureChatOpenAI(
    azure_deployment="gpt-4.1-mini",         # deployment name from Azure portal
    azure_endpoint="https://sinhalkay-7416-resource.cognitiveservices.azure.com/",           # your endpoint
    api_version="2023-12-01-preview",
    api_key="mykey",
    temperature=0.0
)
```

```python

# Tool: Python REPL for heavy math
python_tool = Tool(
    name="Python Calculator",
    func=PythonREPL().run,
    description="Useful for large computations (e.g., powers beyond 20). Input must be valid Python."
)

# Initialize agent
agent = initialize_agent(
    tools=[python_tool],
    llm=llm,
    agent="zero-shot-react-description", # reasoning + tool use
    verbose=True
)
```

```python
# Example query
query = "Calculate 10000 * 12**2 - 12 + log(10)"
```

```python
result = agent.run(query)
print("Final Answer:", result)
```
